```python
import pickle

import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('reconstruction_errors.pickle', 'rb') as f:
    errors = pickle.load(f)
    
    plt.figure()
    plt.plot(0.001 * np.arange(1, 501, 1), errors)
    plt.title('График сходимости по alpha')
    plt.xlabel('alpha')
    plt.ylabel('Ошибка реконструирования')
    plt.savefig('alpha_convergence.pdf', format='pdf', dpi=150)
```
